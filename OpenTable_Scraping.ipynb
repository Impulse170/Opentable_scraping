```
import csv
from requests_html import HTMLSession
from bs4 import BeautifulSoup

i = 1  # first page
output_file = "reviews.csv"

# Open the CSV file in write mode because I suck at this
with open(output_file, "w", newline='', encoding="utf-8") as csvfile:
    # Create a CSV writer object
    csvwriter = csv.writer(csvfile)

    # Write the header row
    csvwriter.writerow(["Reviewer", "Location", "Review", "Date", "Rating"])

    while i <= 106:  # Iterate until page 108
        url = f"https://www.opentable.ca/r/salt-and-ash-halifax?page={i}"
        session = HTMLSession()
        response = session.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        reviews = soup.find_all('li', {'class': "afkKaa-4T28-"})

        for review in reviews:
            reviewer = review.find('p', {'class': "_1p30XHjz2rI- C7Tp-bANpE4-"}).text
            location = review.find('p', {'class': "POyqzNMT21k- C7Tp-bANpE4-"}).text
            d = review.find('p', {'class': "iLkEeQbexGs-"}).text
            date = d[9:]
            scores = review.find_all('li', {'class': "-k5xpTfSXac-"})
            for score in scores:
              rating = score.find('span', {'class': "-y00OllFiMo-"}).text
            rev = review.find('span', {'class': "l9bbXUdC9v0- ZatlKKd1hyc- ukvN6yaH1Ds-"}).text
            print("Reviewer:", reviewer, "\nLocation:", location, "\nReview:", rev,'\nDate', date, "\n---" )

            # Write the review data to the CSV file
            csvwriter.writerow([reviewer, location, rev, date, rating])

        i += 1

print(f"Reviews have been saved to {output_file}")
```

```
import pandas as pd
import numpy as np
import re
```

```
df = pd.read_csv("reviews.csv")
```

```
df.head()
```

```
def clean(text):
   # Remove non-printable characters
    text = re.sub(r'[^\x00-\x7F]', '', text)
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    # Remove any unwanted special characters
    text = re.sub(r'[^\w\s,.!?]', '', text)
    return text
df['Review'] = df['Review'].apply(clean)
```
